```python
import numpy as np

# Dataset
ages = [18, 19, 20, 20, 21, 21, 22, 22, 23, 24, 25, 26, 27, 28, 30, 32, 35, 40, 45, 70]

# Calculate mean and standard deviation
mean = np.mean(ages)
std = np.std(ages)

print(f"Mean: {mean}")
print(f"Standard Deviation: {std}")

# Calculate Z-scores
z_scores = [(x - mean) / std for x in ages]

# Find outliers
outliers = [ages[i] for i in range(len(ages)) if abs(z_scores[i]) > 3]
print(f"Outliers: {outliers}")

# Remove outliers
clean_data = [ages[i] for i in range(len(ages)) if abs(z_scores[i]) <= 3]
print(f"New dataset length: {len(clean_data)}")
```
